Lets Grade the math_score Alphabetically according to a score range for instance (score between 80-100 should be graded 'A' ) 

```python
def Grade(math_score):
    if (math_score >= 80): 
        return 'A'
    if (math_score >= 70): 
        return 'B'
    if (math_score >= 60): 
        return 'C'
    if (math_score >= 50): 
        return 'D'
    if (math_score >= 40): 
        return 'E'
    else:
        return "F"
```

```python
student['math_grade']= student.apply(lambda x : Grade (x['math_score']), axis = 1)
```
